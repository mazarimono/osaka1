## エクセルの概観

- 10行目以降にデータが入っている。      
- 10行目から14行目は過去5年分の数値。そして16行目から39行目は区別の値が記録されている。     
- 49行目から78行目には軽自動車のデータがあるが、最初のところに軽自動車の項目があるため、そこを見る必要はないだろう。      
- データが入っている列は     
- [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 31, 33, 34, 37, 40, 41, 43, 44]
<br>
- 全部作り一緒やんと思ったら違った。   
- H23までは特殊・特種用途者が区別されていなかった。つまり。後の年の方の特種用途車のところをいじる必要があるということだ。     
- 地域別のところはそのままで良いかと思ったら、ちょっと前のデータになると、文字の処理が必要であることが分かった。     
<br>       
- 列のパターンは2種類でなく、3種類であることが分かった。     
- コラム数の変化もH25に一回、H23に一回行われている。

```python
# 自動車種類のコラム
car_osaka = ['総数', '電気自動車総数', '乗用車総数', '普通乗用車', '小型乗用車', '貨物車総数', '貨物普通車', '貨物小型車', '貨物被牽引車',
            'バス', '特種特殊総数', '普通特種', '小型特種', '大型特種', '小型二輪', '軽自動車', '電気軽自動車', '小型特殊自動車', '電気小型特殊自動車']
```

```python
dfH29 = df['H29'].copy()
dfH29 = dfH29.iloc[10:40 , [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 31, 33, 34, 37, 40, 41, 43, 44]]

car_osaka = ['総数', '電気自動車総数', '乗用車総数', '普通乗用車', '小型乗用車', '貨物車総数', '貨物普通車', '貨物小型車', '貨物被牽引車',
            'バス', '特種特殊総数', '普通特種', '小型特種', '大型特種', '小型二輪', '軽自動車', '電気軽自動車', '小型特殊自動車', '電気小型特殊自動車']

car_osaka1 = car_osaka.copy()
car_osaka1.insert(0, '年度')


def make_rowcol(df):
    df.index = [x for x in range(len(df))]    
    df.columns = [x for x in range(len(df.columns))]
    return df

# 行列を番号にする    
dfH29 = make_rowcol(dfH29)

# 全体データと地域データを分ける
dfH29_1 = dfH29.iloc[:5, :]
dfH29_2 = dfH29.iloc[6:, :]

# 全体データの処理
# 全体データからNaN, '平　成', '()'を取り除く
def remove_NaN(df):
    df.iloc[0, 0] = df.iloc[0, 0].replace('平　成　', '')
    df = df.T.dropna()
    df = df.T
    # カッコ、カンマ、マイナスを取り除き、平成を西暦に変える
    problem_col = [2, 17, 19]
    for i in range(len(df)):
        for i2 in problem_col:
            df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(',', '')
            df.iloc[i, i2] = int(df.iloc[i, i2])
            if df.iloc[i, i2] < 0:
                df.iloc[i, i2] = df.iloc[i, i2] * -1
    return df

def heiseireki(df):
    for i in range(len(df)):
        if len(df.iloc[i, 0]) == 2:
            df.iloc[i, 0] = int(df.iloc[i, 0]) + 1988
        else:
            pass
    return df

def matome_zentai(df):
    df = remove_NaN(df)
    df = heiseireki(df)
    df.columns = car_osaka1

    return df
    
    
dfH29_1 = matome_zentai(dfH29_1)

car_osaka2 = car_osaka.copy()
car_osaka2.insert(0, '地域')

# 区別の処理は簡単そう
def kubetsu(df, year):
    df = df.iloc[:, 1:]
    df.columns = car_osaka2
    df['年度'] = year
    df.index = [i for i in range(len(df))]
    return df

dfH29_2 = kubetsu(dfH29_2, 2016)
```
