```python
def make_sheet(df, year):
    df1 = df[year].iloc[10:40,:]
    df1.columns = [i for i in range(len(df1.columns))]
    df1_1 = df1.iloc[:5, :]
    df1_2 = df1.iloc[6:, :]
    return df1_1, df1_2

def s26to29(df, df1):
    df = df.iloc[:,[1,2,5,6,8,11,14,17,20,23,25,28,30,31,33,34,37,40,41,43,44]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,31,33,34,37,40,41,43,44]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    print(df)
    print(df1)
    
def s24to25(df, df1):
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,31,32,33,36,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,31,32,33,36,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1
    
def s21to23(df,df1):
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1

def s20(df,df1):
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,34,37,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,34,37,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1 

def s19(df,df1):
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,37,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,37,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1 

def s15to18(df,df1):
    if df.isnull().iloc[4,36]:
        df.iloc[4,36] = df.iloc[4,37]
        if df.isnull().iloc[4,36]:
            df.iloc[4,36] = 0
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1 

# H14シートのみ
def s14(df,df1): 
    df14[0].iloc[1:5, 1] = df14[0].iloc[1:5, 2]
    df = df.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df = remove_nengo(df)
    df = df.T
    df = df.dropna().T
    df1 = df1.iloc[:, [1,2,5,6,8,11,14,17,20,23,25,28,30,33,36,39,40,42,43]]
    df1 = df1.iloc[:, 1:]
    print(len(df.columns))
    print(len(df1.columns))
    return df, df1     

# シートの1行目の平成を取り除く
def remove_nengo(df):
    df.iloc[0, 0] = df.iloc[0, 0].replace('平　成　', '')  # H29 - H27
    df.iloc[0, 0] = df.iloc[0, 0].replace('平 成  ', '') # H23 - H26
    df.iloc[0, 0] = df.iloc[0, 0].replace('平 成　', '') # H17 - H22
    df.iloc[0, 0] = df.iloc[0, 0].replace('平成', '') # H14 - H16
    return df

# 行、列に番号をふる
def number_count(df, df1):
    df.columns = [i for i in range(len(df.columns))]
    df.index = [i for i in range(len(df.index))]
    df1.columns = [i for i in range(len(df1.columns))]
    df1.index = [i for i in range(len(df1.index))]
    return df, df1

# 数値以外のデータを数値にする
```

```python
def test_case(df, year):
    df,df1 = make_sheet(df, year)
    df, df1 = s14(df, df1)
    df, df1 = number_count(df, df1)
    print(df, df1)
```
