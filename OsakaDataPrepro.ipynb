```python
# 年度のエクセルファイルを必要部分を切り出す
def seikei(df):
    df = df.copy()
    osa_car1 = [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 31, 33, 34, 37, 40, 41, 43, 44]
    osa_car2 = [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 31, 32, 33, 36, 39, 40, 42, 43]
    osa_car3 = [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 33, 36, 39, 40, 42, 43]
    osa_car4 = [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 34, 37, 39, 40, 42, 43]
    osa_car5 = [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 33, 36, 39, 40, 42, 43]
    if len(df.columns) == 47:
        df1 = df.iloc[10:40, osa_car1]
        df2 = df1.iloc[:5, :]
        df2 = df2.T.dropna()
        df2 = df2.T
        if len(df2.columns) != 20:
            df1 = df.iloc[10:40, osa_car2]
    elif len(df.columns) == 46:
        df1 = df.iloc[10: 40, osa_car3]
        df2 = df1.iloc[:5, :]
        df2 = df2.T.dropna()
        df2 = df2.T
        if len(df2.columns) < 18:
            df1 = df.iloc[10:40, osa_car4]
    else:
        pass
    print(df1)
    return pd.DataFrame(df1)

def bunri(df):
    df_all = df.iloc[:5, :]
    df_area = df.iloc[6:, :]
    return pd.DataFrame(df_all), pd.DataFrame(df_area)

def remove_n_replace(df, column_name1, column_name2):
    df.iloc[0, 0] = df.iloc[0, 0].replace('平　成　', '')
    df.iloc[0, 0] = df.iloc[0, 0].replace('平 成  ', '')
    df.iloc[0, 0] = df.iloc[0, 0].replace('平 成　', '')
    df.iloc[0, 0] = df.iloc[0, 0].replace('平成', '')
    df = df.T.dropna()
    df = df.T
    if len(df.columns) == 20:
        problem_col= [2, 17, 19]
        df.columns = column_name1
        for i in range(len(df)):
            for i2 in problem_col:
                df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
                df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
                df.iloc[i, i2] = int(df.iloc[i, i2].replace(',', ''))
                if df.iloc[i, i2] < 0:
                    df.iloc[i, i2] = df.iloc[i, i2] * -1
        
        df['特種用途車'] = df['普通特種'] + df['小型特種']
        df = df.drop(['特種特殊総数', '普通特種', '小型特種'], axis=1)
        return df
    
    elif len(df.columns) == 18:
        problem_col = [2, 15, 17]
        df.columns = column_name2
        for i in range(len(df)):
            for i2 in problem_col:
                df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
                df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
                df.iloc[i, i2] = int(df.iloc[i, i2].replace(',', ''))
                if df.iloc[i, i2] < 0:
                    df.iloc[i, i2] = df.iloc[i, i2] * -1
        return df
    else:
        print(len(df.columns))

# 平成を西暦に(データフレーム用)
def heiseireki(df):
    print(df)
    for i in range(len(df)):
        if len(str(df.iloc[i, 0])) == 2:
            df.iloc[i, 0] = int(df.iloc[i, 0]) + 1988
        else:
            print(df.iloc[i, 0])
    return df

# 平成を西暦に（単独用）
def heiseireki_alone(year):
    year = int(year.replace('H', ''))
    year = year + 1988
    return year

def remove_area(df, column_name1, column_name2):
    df = df.iloc[:, 1:].copy()
    if len(df.columns) == 20:
        problem_col=[2, 17, 19]
        df.columns = column_name1
        for i in range(len(df)):
            for i2 in problem_col:
                df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
                df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
                try:
                    df.iloc[i, i2] = int(df.iloc[i, i2].replace(',', ''))
                except:
                    df.iloc[i, i2] = 0
                    
                if df.iloc[i, i2] < 0:
                    df.iloc[i, i2] = df.iloc[i, i2] * -1
                elif df.iloc[i, i2] == 'NaN':
                    df.iloc[i, i2] = 0
                else:
                    pass
                
        df['特種用途車'] = df['普通特種'] + df['小型特種']
        df = df.drop(['特種特殊総数', '普通特種', '小型特種'], axis=1)
        return df
    
    elif len(df.columns) == 18:
        problem_col=[2, 15, 17]
        df.columns = column_name2
        for i in range(len(df)):
            for i2 in problem_col:
                df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
                df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
                try:
                    df.iloc[i, i2] = int(df.iloc[i, i2].replace(',', ''))
                except:
                    df.iloc[i, i2] = 0
                    
                if df.iloc[i, i2] < 0:
                    df.iloc[i, i2] = df.iloc[i, i2] * -1
                elif df.iloc[i, i2] == 'NaN':
                    df.iloc[i, i2] = 0
                else:
                    pass
        return df
    
def area_betsu(df, year, column_name1, column_name2):
    df = remove_area(df, column_name1, column_name2)
    df['年度'] = year
    df.index = [i for i in range(len(df))]
    return df
```

```python
# 年度のエクセルファイルから必要部分だけ切り出す
def seikei(df):
    df = df.copy()
    df = df.iloc[10:40 , [1, 2, 5, 6, 8, 11, 14, 17, 20, 23, 25, 28, 30, 31, 33, 34, 37, 40, 41, 43, 44]]
    return df

# 全体情報と地域情報に分離する
def bunri(df):
    df_all = df.iloc[:5, :]
    df_area = df.iloc[6:, :]
    return df_all, df_area

# 全体情報を整理する。まずは除去と置き換え
def remove_n_replace1(df):
    df.iloc[0, 0] = df.iloc[0, 0].replace('平　成　', '')
    df = df.T.dropna()
    df = df.T
    print(df)
    # カッコ、カンマ、マイナスを取り除く
    problem_col = [2, 17, 19]
    for i in range(len(df)):
        for i2 in problem_col:
            df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(',', '')
            df.iloc[i, i2] = int(df.iloc[i, i2])
            if df.iloc[i, i2] < 0:
                df.iloc[i, i2] = df.iloc[i, i2] * -1
    return df

# H23以前
def remove_n_replace2(df):
    df.iloc[0, 0] = df.iloc[0, 0].replace('平　成　', '')
    df = df.T.dropna()
    df = df.T
    print(df)
    # カッコ、カンマ、マイナスを取り除く
    problem_col = [2, 15, 17]
    for i in range(len(df)):
        for i2 in problem_col:
            df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(',', '')
            df.iloc[i, i2] = int(df.iloc[i, i2])
            if df.iloc[i, i2] < 0:
                df.iloc[i, i2] = df.iloc[i, i2] * -1
    return df

# 平成を西暦に(データフレーム用)
def heiseireki(df):
    for i in range(len(df)):
        if len(str(df.iloc[i, 0])) == 2:
            df.iloc[i, 0] = int(df.iloc[i, 0]) + 1988
        else:
            pass
    return df

# 平成を西暦に（単独用）
def heiseireki_alone(year):
    year = int(year.replace('H', ''))
    year = year + 1988
    return year

def zentai1(df, column_name):
    df = remove_n_replace1(df)
    df = heiseireki(df)
    df.columns = column_name
    return df

def zentai2(df, column_name):
    df = remove_n_replace2(df)
    df = heiseireki(df)
    df.columns = column_name
    return df    

# 地域データカッコなど処理
def remove_area1(df):
    problem_col= [2, 17, 19]
    for i in range(len(df)):
        for i2 in problem_col:
            df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(',', '')
            df.iloc[i, i2] = int(df.iloc[i, i2])
            if df.iloc[i, i2] < 0:
                df.iloc[i, i2] = df.iloc[i, i2] * -1
            elif df.iloc[i, i2] == 'NaN':
                df.iloc[i, i2] = 0
            else:
                pass
    return df

def remove_area2(df):
    problem_col= [2, 15, 17]
    for i in range(len(df)):
        for i2 in problem_col:
            df.iloc[i, i2] = str(df.iloc[i, i2]).replace('(', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(')', '')
            df.iloc[i, i2] = df.iloc[i, i2].replace(',', '')
            df.iloc[i, i2] = int(df.iloc[i, i2])
            if df.iloc[i, i2] < 0:
                df.iloc[i, i2] = df.iloc[i, i2] * -1
            elif df.iloc[i, i2] == 'NaN':
                df.iloc[i, i2] = 0
            else:
                pass
    return df

# 地域データの整頓
def area_betsu1(df, year, column_name):
    df = df.iloc[:, 1:]
    df = remove_area1(df)
    df.columns = column_name
    df['年度'] = year
    df.index = [i for i in range(len(df))]
    return df

def area_betsu2(df, year, column_name):
    df = df.iloc[:, 1:]
    df = remove_area2(df)
    df.columns = column_name
    df['年度'] = year
    df.index = [i for i in range(len(df))]
    return df    
```
