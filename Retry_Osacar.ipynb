```python
def excel_to_csv(file_path='osaka_car.xls'):
    excel = pd.ExcelFile(file_path)
    df = pd.read_excel('osaka_car.xls', sheet_name = excel.sheet_names)
    for i in excel.sheet_names:
        i2 = int(i.replace('H', ''))
        if i2 >= 26:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s26to29(df1, df2)
            df1 = make_cols_all(df1, [2,17,19])
            df2 = make_cols_all(df2, [2, 9, 17, 19])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 >= 24:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s24to25(df1, df2)
            df1 = make_cols_all(df1, [2,17,19])
            df2 = make_cols_all(df2, [2, 9, 17, 19])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 >= 21:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s21to23(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17]) 
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 == 20:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s20(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 == 19:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s19(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 == 18:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s18(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 >= 15:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s15to17(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
        elif i2 == 14:
            df1, df2 = make_sheet(df, i)
            df1, df2 = s14(df1, df2)
            df1 = make_cols_all(df1, [2,15,17])
            df2 = make_cols_all(df2, [2, 9, 15, 17])
            df1.to_csv('./csv1/all_{}.csv'.format(i))
            df2.to_csv('./csv1/kubetsu_{}.csv'.format(i))
```
